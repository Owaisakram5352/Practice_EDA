```python
df1 = df
df1.drop("policy_id",axis=1)
```

### Creating box plot with respect to target value to find outliers

```python
num_col = [col for col in df1.columns if df1[col].dtype != "object"]
for col in num_col:
  plt.figure()
  sns.boxplot(x="is_claim",y=df1[col],data = df1)
  plt.xlabel("is_claim")
  plt.ylabel(col)
  plt.title('Box Plot for ' + col)

plt.show()
```

### Checking for Correlation and creating heat map

```python
df1.corr()
```

```python
correlation_matrix = df1.corr()
plt.figure(figsize = (17,17))
sns.heatmap(correlation_matrix,annot= True,cmap = "Pastel1")
plt.title("Correlation matrix of the dataset")
plt.show()
```

### Creating a count plot for the visualizations

```python
for col in num_col:
  plt.figure(figsize=(20,5))
  sns.countplot(x=df1[col],data = df1)
  plt.xlabel(col)
  plt.ylabel("Count")
  plt.title("Count plot of " + col)

plt.show()
```

### Extracting the categorical columns

```python
catfeat= [col for col in df1.columns if df1[col].dtype == "object" and col not in ["policy_id"]]
features = [col for col in df1.columns if col not in ["policy_id","is_claim"]]
target = [col for col in df1.columns if col=="is_claim"]
```

```python
print(catfeat)
print(len(catfeat))
print(features)
print(len(features))
print(target)
```

### Splitting features and target columns

```python
x,y =df1.loc[:,features],df1.loc[:,target]
print(x.shape)
```

### Label encoding the categorical columns

```python
labelEncode = LabelEncoder()
for col in catfeat:
  x[col]=labelEncode.fit_transform(df1[col])
```
